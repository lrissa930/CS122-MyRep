#### What is the main difference between for-loop and while-loop?
#### In which case the while-loop has advantage?

A for-loop is used when you know exactly how many times you want to repeat something. A while-loop is used when you want to repeat until a condition changes.

For-Loop
- Best when number of iterations is known in advance
- Uses a sequence such as range(), a list, or a string
-Example of looping 10 times below
> for i in range(10):
  print(i)

While-Loop
- Best when the number of iterations is unknown
- Continues until a condition becomes false
- Example of asking for input until the user enters a number > 100 below
> num = 0
  while num <= 100:
    num = int(input('Enter a number: ;))

A while-loop is better when:
- You don't know how many times the loop will run
- You are waiting for a condition to change
- You are processing user input until they meet a requirement

##### ✓ No need to use the "list" data type.

````python
for i in range(5):
    rdnum = random.randint(0,10)
    if i == 0: # if it is the first iteration
        maxval = rdnum # We assume that the first rdnum is greatest number
    if rdnum > maxval:  # if the next rdnum is greater than the current maxval, keep it as maxval
        maxval = rdnum
    print (rdnum, end=' ')

print (f'\nThe greatest random number is {maxval}')
````

#### Follow the algorithm steps to find the least number
> 1) Assume that the 1st element is the least number    
>> <code> minval = mylist[0] </code>
> 2) Iterate all elements from 2nd to the last
>> <code> for i in range(len(mylist)): </code>
> 3) If any element in the list is less than the current, replace the minval with the current element

````python
mylist = [10, 20, 30, 40, 50]
minval = mylist[0]

for i in range(len(mylist)):
  if mylist[i] < minval:
    minval = mylist[i]

print(f'The least number is {minval}')
````

#### <p style="font-family: Arial; font-size:1.4em; background-color: yellow; color:black;"> Problem #6.5  </p>
#### Write a code using for-else to find if any duplicate exists in the list
#### If a duplicate is found, print it and stop. If no duplicate, print 'No duplicates'.

#### Explain why the `else` block executes or doesn't for each input.

The `else` block executes:

when a duplicate is found:
* the inner loop hits `break`
* skips the inner loop's `else`
* outer loop hits its own `break`
* `else` does not run

when no duplicate exists:
* inner loop never breaks -> `else` runs each time
* outer loop finishes normally (no break)
* outer loop's `else` executes

````python
mylist = [10, 20, 30, 40, 50, 20]

for i in range(len(mylist)):
  for j in range(i + 1, len(mylist)):
    if mylist[i] == mylist[j]:
      print(f'Duplicate found: {mylist[i]}')
      break

  else:
    continue
  break
else:
  print(f'No duplicates.')
````

#### <p style="font-family: Arial; font-size:1.4em; background-color: yellow; color:black;"> Problem #6.7  </p>
#### Write a code using nested for-else to search for a target value in a 2D list (matrix)
#### Print 'Found' if the target exists, 'Not found' otherwise.

````python
#### <p style="font-family: Arial; font-size:1.4em; background-color: yellow; color:black;"> Problem #6.7  </p>
#### Write a code using nested `for`-`else` to search for a target value in a 2D list (matrix)
#### Print "Found" if the target exists, "Not found" otherwise.
#### **Requirement:** use `break` and `else` — the outer `else` should only print "Not found" if the target was never found in any row.

matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
target = 5

for row in matrix:
  for value in row:
    if value == target:
      print('Found')
      break

  else:
    continue
  break

else:
  print('Not found')
````
