# <p style="font-family: Arial; font-size:1.4em;color:red;"> 7. Complex Compound Conditions </p>

#### Venn Diagram

> Given the Universe Set U and three sets A, B, C with the following relationships:
> - A ⊂ B (A is a subset of B)
> - A ∩ C = ∅ (A does NOT intersect C)
> - B ∩ C ≠ ∅ (B intersects C)

![Venn Diagram](venn_diagram.png)

#### <p style="font-family: Arial; font-size:1.4em; background-color: yellow; color:black;"> Problem #7.1 </p>

#### Create your own Venn diagram scenario that satisfies:
> - A ⊂ B (A is a subset of B)
> - A ∩ C = ∅ (A does NOT intersect C)
> - B ∩ C ≠ ∅ (B intersects C)

#### Example: A = Sparkling Water, B = Drinks, C = Hot
> - A ⊂ B: Sparkling water is a drink
> - A ∩ C = ∅: Sparkling water is not hot
> - B ∩ C ≠ ∅: Some drinks are hot (coffee, tea)
